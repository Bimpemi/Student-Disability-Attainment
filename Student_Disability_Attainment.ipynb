### Requirement FR13 - Test if there is any difference between the attainment of disabled and non-disabled students

```python
#Using an appropriate Python library, test if there is any statistically significant 
#difference between the attainment of disabled and non-disabled students. 
# Separate the data into two groups based on whether the student is disabled or not

import scipy.stats as stats 


# Create two groups of data, one for disabled and one for non-disabled students
disabled_students = clean_records['disability']=='Y'
non_disabled_students = clean_records['disability']=='N'
attainment= clean_records['score']
#The t-test was used because it comparing a categorical data and continuous data
t_test_disabled, p_value_disabled = stats.ttest_1samp(disabled_students,attainment)
#corr_disabled, p_value_disabled = scipy.stats.pearsonr(disabled_students,attainment)
print('t-test of disability students=',t_test_disabled)
print('p_value of disability students=',p_value_disabled)

print('..........................................................................')

t_test_non_disabled, p_value_non_disabled = stats.ttest_1samp(non_disabled_students,attainment)
print('t-test of non-disability student=',t_test_non_disabled)
print('p_value of non-disability student=',p_value_non_disabled)
```

```python
#information about students with disability
disabled_students.info()
disabled_students.describe()
```

```python
#information about students without disability
non_disabled_students.info()
non_disabled_students.describe()
```

Explaination about the statistical significant difference between the attainment of disabled and non-disabled students

The null hypothesis (Ho) is that there is no difference between the attainment of disabled and non-disabled students.
To reject or not reject the null hypothesis, The t-test and the P-value were computed. 
The P-value which is the probability value used in statistics to reject or not reject the null hypothesis is when P<0.05, therefore, if the P-value<0.05, it is statistically significant, and the null hypothesis is rejected.

In the above result the p-value to test if there is any statistically significant difference 
between disabled and non-disabled students was 0.0 respectively, which means the P-value 
is not <0.05, therefore the null hypothesis is not rejected and the alternative hypothesis (Ha) which is there is difference between the attainment of disabled and non-disabled students is rejected.

In conclusion, there is a statistically significant no difference between the attainment of disabled and non-disabled students.
